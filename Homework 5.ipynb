```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# 1) Generate a synthetic dataset: Normal(0,1)
np.random.seed(42)
n = 30
x = np.random.normal(loc=0, scale=1, size=n)

# ------------------------------------------------------------------------
# Model 1: theta ~ Normal(0,1), tau ~ Gamma(alpha=2, beta=1)
# ------------------------------------------------------------------------
with pm.Model() as model1:
    # Define priors for theta and tau
    theta = pm.Normal("theta", mu=0, sigma=1)
    tau = pm.Gamma("tau", alpha=2, beta=1)
    
    # Likelihood: Normal with mean theta and precision tau
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    
    # MCMC sampling
    idata1 = pm.sample(
        draws=1000, tune=1000, chains=2, 
        return_inferencedata=True, target_accept=0.9, random_seed=42
    )

# Diagnostics: trace plots and summary
az.plot_trace(idata1)
plt.suptitle("Model 1: Normal-Gamma Priors", y=1.02)
plt.tight_layout()
plt.show()
print("Model 1 Summary:")
print(az.summary(idata1, round_to=2), "\n\n")
```

```python
import pymc as pm
import numpy as np
import arviz as az
import matplotlib.pyplot as plt

# 1) Generate a synthetic dataset: Normal(0,1)
np.random.seed(42)
n = 30
x = np.random.normal(loc=0, scale=1, size=n)
# ------------------------------------------------------------------------
# Model 3: theta ~ Laplace(mu=0, b=1), tau ~ Beta(alpha=2, beta=2)
# ------------------------------------------------------------------------
with pm.Model() as model3:
    # Define priors for theta and tau
    theta = pm.Laplace("theta", mu=0, b=1)
    tau = pm.Beta("tau", alpha=2, beta=2)
    
    # Likelihood
    x_obs = pm.Normal("x_obs", mu=theta, tau=tau, observed=x)
    
    # MCMC sampling
    idata3 = pm.sample(
        draws=1000, tune=1000, chains=2, 
        return_inferencedata=True, target_accept=0.9, random_seed=42
    )

# Diagnostics: trace plots and summary
az.plot_trace(idata3)
plt.suptitle("Model 3: Laplace-Beta Priors", y=1.02)
plt.tight_layout()
plt.show()

print("Model 3 Summary:")
print(az.summary(idata3, round_to=2), "\n\n")
```

<p><strong>Slice Sampling Algorithm:</strong></p>
<ul>
  <li><em>Target Function</em>: We have <code>f</code> (unnormalized) as our distribution’s PDF.</li>
  <li><em>Vertical Draw</em>: From the current state <code>x</code>, choose a uniform <code>y</code> in <code>[0, f(x)]</code>.</li>
  <li><em>Horizontal Slice</em>: Find the interval where <code>f(x') &gt; y</code>, then draw a new candidate <code>x'</code> uniformly in that region. 
      If <code>f(x')</code> remains &gt; y, accept it. Otherwise, shrink the interval and repeat.</li>
  <li><em>Markov Chain Update</em>: <code>x'</code> becomes the new state in the chain, forming a valid MCMC update.</li>
</ul>

<p><strong>How the Code Works:</strong></p>
<ol>
  <li><code>slice_f_at_y</code> locates the interval <code>[a,b]</code> where <code>f(x_grid)</code> &gt; <code>y</code>. 
      It proposes a new point <code>x_</code> uniformly in <code>[a,b]</code>. 
      If <code>f(x_)</code> &gt; <code>y</code>, it returns immediately. Otherwise, it narrows the bracket and calls <code>slice_f_at_y_</code> for further refinement.</li>
  <li><code>slice_f_at_y_</code> is the recursive helper function that repeatedly shrinks the bracket <code>[x_l, x_r]</code> until a valid <code>x_</code> is found.</li>
  <li>In the main loop, each iteration draws <code>y = Uniform(0, f(x[t,0]))</code>, then calls <code>slice_f_at_y</code> to find the new sample <code>x[t+1,0]</code>. 
      We also record the number of tries in <code>x[t+1,2]</code> for reference.</li>
  <li>The code generates samples from a <strong>Beta(1.5,3)</strong> distribution, plots a histogram of the chain’s <code>x</code> values, and demonstrates how slice sampling converges to the desired distribution.</li>
</ol>

<p><strong>Replacing Metropolis–Hastings in Gibbs:</strong></p>
<ul>
  <li>In a Metropolis within Gibbs approach, each parameter update often requires an accept/reject step. 
      If the conditional densities are only known up to a constant, slice sampling can directly sample from the conditional “slice” without needing an explicit acceptance ratio.</li>
  <li>We pick the current parameter value as the starting <code>x</code>, draw <code>y</code> in <code>[0, f(x)]</code>, and sample a new <code>x</code> from the region <code>f(x') &gt; y</code>. 
      This fully respects the conditional distribution without needing a normalizing factor.</li>
</ul>

<p>Thus, slice sampling is an elegant, self-tuning technique that can simplify MCMC, often yielding better mixing and eliminating the need for complex proposal distributions.</p>
